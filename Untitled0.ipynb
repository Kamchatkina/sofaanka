Задание 1

```python
import pandas as pd #импортируем нужную библиотеку
```

```python
df = pd.read_csv('Вариант 1-2.csv') #чтение файла
```

```python
df['color'].unique() #проверяем на ошибки в столбце цвет
```

```python
df['color'] = df['color'].str.strip() #избавляемся от ошибок в столбце с обозначением цвета
df.loc[df['color'] == 'blu', 'color'] = 'blue'
df['color'].unique()
```

```python
df['brand'].unique() #проверяем на ошибки в столбце бренд
```

```python
df['brand'] = df['brand'].str.title() #избавляемся от этих ошибок
df.loc[df['brand'] == 'Bmv'] = 'Bmw'
df.loc[df['brand'] == 'Hyunday'] = 'Hyundai'
df['brand'].unique()
```

```python
df['model'].unique() #есть ошибки с Hyundai и Bmw
```

```python
df[(df['model'] == 'Hyundai')|(df['model'] == 'Bmw')] #данные неверно заполнены
```

```python
df.loc[(df['gasoline_consumption'] == 'Bmw') |
       (df['gasoline_consumption'] == 'Hyundai')|
        (df['gasoline_consumption'] == 'Null'),
        ['person', 'color', 'brand', 'model', 'price', 'discount', 'date', 'gasoline_consumption']] = ['', '', '', '', '', '', '', ''] #избавляемся от ошибок с Bmw and Hyndai, так как у них не правильно указаны даты цены и т.д.
```

```python
df.duplicated().sum()#считаем дубликаты
```

```python
df = df.drop_duplicates()#удаляем дубликаты
```

```python
df = df.drop(labels=[28])#удаляем пустую строку
```

Задание 2

```python
#средняя цена по отобранным брендам за нужный период и их количество
df['price'] = df['price'].astype(float)
df['new_price'] = df['price'] - df['price']*df['discount']/100
new_df = df[df['brand'] != 'Lada']
filter = new_df[(new_df['date'] >= '2021-01-01') & (new_df['date'] < '2022-01-01')]
avg_price = filter['new_price'].mean()
filtered_df = filter.loc[filter['new_price'] > avg_price]
print(round(filtered_df['new_price'].mean(), 2), len(filtered_df))
```

Задание 3

```python
#топ 5 цветов автомобилей по количеству заказов
gr = df.groupby('color').agg(popular_color = ('person', 'count')).reset_index()
sort_df = gr.sort_values(by='popular_color', ascending=False)
popular_colors = sort_df.head().reset_index(drop = True)
popular_colors
```

```python
#средняя цена и общая стоимость продаж для каждого отобранного бренда
filtered_df = df[df['color'].isin(popular_colors['color'])]
result = filtered_df.groupby('brand', as_index=False).agg(avg_car = ('new_price', 'mean'),
                                                          sum_car = ('new_price', 'sum'))
result[['avg_car', 'sum_car']] = result[['avg_car', 'sum_car']].astype(float)
result['sum_car'] = result['sum_car'].round(2)
result['avg_car'] = result['avg_car'].round(2)
result
```

Задание 4


```python
#создаем сводную таблицу по годам и брендам
df['years'] = pd.to_datetime(df['date']).dt.year
quarter_total = df.groupby(['brand', 'years'], as_index=False).agg(sum_cars = ('new_price', 'sum'))
quarter_total['sum_cars'] = quarter_total['sum_cars'].astype(float)
quarter_total['sum_cars'] = quarter_total['sum_cars'].round(2)
quarter_total
```

Графики

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=[14,5])
#построение графика номер 1
plt.subplot(1, 2, 1)
filter_Tesla = quarter_total[quarter_total['brand']=='Tesla']
filter_Porsche = quarter_total[quarter_total['brand']=='Porsche']
filter_Lada = quarter_total[quarter_total['brand']=='Lada']
filter_Kia = quarter_total[quarter_total['brand']=='Kia']
filter_Hyundai = quarter_total[quarter_total['brand']=='Hyundai']
filter_Bmw = quarter_total[quarter_total['brand']=='Bmw']
filter_Audi = quarter_total[quarter_total['brand']=='Audi']
plt.plot(filter_Tesla['years'], filter_Tesla['sum_cars'], marker='o', label='Tesla')
plt.plot(filter_Tesla['years'], filter_Porsche['sum_cars'], marker='o', label='Porsche')
plt.plot(filter_Tesla['years'], filter_Lada['sum_cars'], marker='o', label='Lada')
plt.plot(filter_Tesla['years'], filter_Kia['sum_cars'], marker='o', label='Kia')
plt.plot(filter_Tesla['years'], filter_Hyundai['sum_cars'], marker='o', label='Hyundai')
plt.plot(filter_Tesla['years'], filter_Bmw['sum_cars'], marker='o', label='Bmw')
plt.plot(filter_Tesla['years'], filter_Audi['sum_cars'], marker='o', label='Audi')
plt.title('Продажи брендов машин в разные года')
plt.legend()
plt.ylim(0, 2200000)
plt.xlabel('Года')
plt.ylabel('Продажи, млн')
#построение графика номер 2
plt.subplot(1, 2, 2)
d = quarter_total.pivot_table(values = 'sum_cars',
                              index = 'brand',
                              columns = 'years',
                              aggfunc = 'sum')
a = plt.imshow(d, cmap = 'ocean')
plt.colorbar(a)
plt.xticks(ticks = range(len(d.columns)), labels = d.columns, rotation = 30)
plt.yticks(ticks = range(len(d.index)), labels = d.index)
plt.title('Продажи брендов машин в разные года')
plt.show()
```

Из графиков можно заметить, что автомобиль Тесла выделяется среди остальных и больше всех принес выручки, а пик продаж произошел в 2018 году. На втором месте идет Bmw и уже оставшиеся бренды машин
